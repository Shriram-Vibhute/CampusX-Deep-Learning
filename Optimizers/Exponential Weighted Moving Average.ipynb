```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DailyDelhiClimate.csv').loc[:, ['date', 'meantemp']]
df.head()
```

```python
plt.figure(figsize = (15, 6))
plt.scatter(df['date'], df['meantemp'])
plt.show()
```

`alpha` = 1 - beta

```python
df['ewma_0.1'] = df['meantemp'].ewm(alpha = 0.9).mean()
df['ewma_0.5'] = df['meantemp'].ewm(alpha = 0.5).mean()
df['ewma_0.9'] = df['meantemp'].ewm(alpha = 0.1).mean()
```

```python
plt.figure(figsize = (15, 6))
plt.scatter(df['date'], df['meantemp'])
plt.plot(df['date'], df['ewma_0.1'], color = 'red')
plt.show()
```

```python
plt.figure(figsize = (15, 6))
plt.scatter(df['date'], df['meantemp'])
plt.plot(df['date'], df['ewma_0.5'], color = 'red')
plt.show()
```

```python
plt.figure(figsize = (15, 6))
plt.scatter(df['date'], df['meantemp'])
plt.plot(df['date'], df['ewma_0.9'], color = 'red')
plt.show()
```

`Observation`

> "When the beta value is high, the Exponential Weighted Moving Average (EWMA) curve appears smooth, whereas a low beta value results in a more fluctuating and sensitive curve."

The **beta** value in Exponential Weighted Moving Average (EWMA) controls **how much we trust the past** compared to the most recent value.

* When **beta is high** (like 0.9 or 0.99), we give **more importance to the past values** and **less to the new values**. So the average changes **slowly**, making the curve **smooth**.

* When **beta is low** (like 0.1 or 0.3), we give **more weight to the new values**, and the past is almost ignored. So the average changes **quickly**, making the curve **fluctuate a lot**.

Think of it like this:

If you’re slowly turning a steering wheel (high beta), the car moves smoothly. If you keep making sharp, fast turns (low beta), the car swerves around a lot.
